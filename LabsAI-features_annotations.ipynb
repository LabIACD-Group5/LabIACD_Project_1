```python
import pylidc as pl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
path = "C:/Users/35192/Desktop/LIDC-IDRI"
f = open('C:/Users/35192/pylidc.conf', 'w')
f.write(f'[dicom]\npath = {path}\n\n')
f.close()
```

```python
#codigo para ter a moda de uma lista
def most_frequent(List):
    unique, counts = np.unique(List, return_counts=True)
    index = np.argmax(counts)
    return unique[index]
```

```python
#Os nomes para as colunas foram obtidos a partir das anotacoes da imagens
panda_anotations = pd.DataFrame(columns= [
                            'id', 
                            'subtlety', 
                            'internalStructure', 
                            'calcification', 
                            'sphericity', 
                            'margin', 
                            'lobulation', 
                            'spiculation', 
                            'texture', 
                            'diameter',
                            'surface_area',
                            'volume','malignancy'])
```

No codigo abaixo fazemos que para cada cluster de anotacoes,calculamos a media ou o valor mais frequente das features,dependente do tipo da feature.

```python
root="C:/Users/35192/Desktop/LIDC-IDRI"
c=0
dirlist = [ item for item in os.listdir(root) if os.path.isdir(os.path.join(root, item)) ]
for di in dirlist:
    sca = pl.query(pl.Scan).filter(pl.Scan.patient_id == di).all()
    count=0
    for scan in sca:
        nods = scan.cluster_annotations()
        for n in range(len(nods)):
            coluna =dict((col, "") for col in panda_anotations.columns)
            cluster= nods[n]
            subtlety=[]
            internalStructure=[]
            calcification=[]
            sphericity=[]
            margin=[]
            lobulation=[]
            spiculation=[]
            texture=[]
            malignancy=[]
            diameter=0
            surface_area=0
            volume=0
            for n in cluster:
                a =pl.query(pl.Annotation).filter(pl.Annotation.id==n.id).first()
                subtlety.append(a.subtlety)
                internalStructure.append(a.internalStructure)
                calcification.append(a.calcification)
                sphericity.append(a.sphericity)
                margin.append(a.margin)
                lobulation.append(a.lobulation)
                spiculation.append(a.spiculation)
                texture.append(a.texture)
                diameter+=a.diameter
                surface_area+=a.surface_area
                volume+=a.volume
                #queremos transformar a malignancy numa classe binaria onde >3 é que aprecenta malignancy e que <3 nao apresenta
                if(a.malignancy>3):
                    malignancy.append(1)
                elif(a.malignancy<3):
                    malignancy.append(0)
                else:
                    malignancy.append(3)
                    
            coluna['id']= f'{di}-{count}'
            coluna['subtlety'] = most_frequent(subtlety)
            coluna['internalStructure'] = most_frequent(internalStructure)
            coluna['calcification'] = most_frequent(calcification)
            coluna['sphericity'] = most_frequent(sphericity)
            coluna['margin'] = most_frequent(margin)
            coluna['lobulation'] = most_frequent(lobulation)
            coluna['spiculation'] = most_frequent(spiculation)
            coluna['texture'] = most_frequent(texture)
            coluna['diameter'] = diameter/len(cluster)
            coluna['surface_area'] = surface_area/len(cluster)
            coluna['volume'] = volume/len(cluster)
            coluna['malignancy'] = most_frequent(malignancy)
            count+=1
            #se o valor mais comun for 3,ou seja,nao sabemos se é ou nao prejudicial,nao adicionamos ao dataset
            if(coluna['malignancy']!=3):
                panda_anotations = panda_anotations.append(coluna, ignore_index=True)
    

panda_anotations.to_csv('annotations_features.csv', sep=',', index=False)

                    
                    
                    
                    
                
```
